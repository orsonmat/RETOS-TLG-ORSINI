```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk

#Cargar base de datos de empleados
EmpleadosAttrition = pd.read_csv('/content/EMPLEADOS.csv')

EmpleadosAttrition.head(5)
```

```python
# Eliminación de variables (columnas) con información irrelevante
# EmployeeCount , EmployeeNumber , Over18, StandardHours

EmpleadosAttrition = EmpleadosAttrition.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis=1)

EmpleadosAttrition.head()
```

```python
# Calcular el tiempo total del empleado en la compañia

# Calcular año de contratación a partir de la fecha exacta de contratación

EmpleadosAttrition['Año'] = pd.to_datetime(EmpleadosAttrition['HiringDate']).dt.year.astype(int)
```

```python
#Validar la posición con error
EmpleadosAttrition['HiringDate'].iloc[229]
```

```python
# Ajustar el error en los datos
EmpleadosAttrition['HiringDate'].iloc[229]= '2/28/2012'
```

```python

# Calcular nuevamente el año de contratación a partir de la fecha exacta de contratación

EmpleadosAttrition['Año'] = pd.to_datetime(EmpleadosAttrition['HiringDate']).dt.year.astype(int)
EmpleadosAttrition['Año']. head()
```

```python

# Calcular los años de antiguedad en la empresa hasta el 2018 (incluido el 2018)
EmpleadosAttrition['YearsAtCompany'] = 2018 - EmpleadosAttrition['Año'] + 1
EmpleadosAttrition['YearsAtCompany'].head()
```

```python
# Cambiar nombre de la columna DistanceFromHome a DistanceFromHome_km

EmpleadosAttrition = EmpleadosAttrition.rename(columns={'DistanceFromHome': 'DistanceFromHome_km'})
EmpleadosAttrition.head()
```

```python
# Eliminar letras km de la columna DistanceFromHome_km

EmpleadosAttrition['DistanceFromHome_km'] = EmpleadosAttrition['DistanceFromHome_km'].astype(str).str.replace('km','').str.strip().astype(int)
EmpleadosAttrition['DistanceFromHome_km'].head()
```

```python
#Eliminar columnas Año y HiringDate pues ya no son necesarias

EmpleadosAttrition = EmpleadosAttrition.drop(['Año', 'HiringDate'], axis=1)
```

```python
#Revisar estructura de la informacón despues de todos los cambios
EmpleadosAttrition.info()
```

```python
# Se han detectado valores nulos en el estatus marital, se reemplazaran por la moda de los datos

EmpleadosAttrition['MaritalStatus']= EmpleadosAttrition['MaritalStatus'].fillna(EmpleadosAttrition['MaritalStatus'].mode()[0])
EmpleadosAttrition['MaritalStatus'].info()
```

```python
# Dataframe con sueldo promedio por departamento

# Agrupar por departamento y calcular el promedio del ingreso mensual
SueldoPromedio = EmpleadosAttrition.groupby('Department')['MonthlyIncome'].mean().reset_index()

# Renombrar las columnas para mayor claridad
SueldoPromedio = SueldoPromedio.rename(columns={'MonthlyIncome': 'PromedioSueldoMensual'})

# Mostrar el nuevo DataFrame
SueldoPromedio
```

```python
# Escalado de variable MonthlyIncome mediante normalización min-max

from sklearn import preprocessing

# Objeto para escalar con MinMax
scaler = preprocessing.MinMaxScaler()

# Transformar la columna MonthlyIncome con método MinMax
EmpleadosAttrition['MonthlyIncome_scaled'] = scaler.fit_transform(EmpleadosAttrition[['MonthlyIncome']])
```

```python
# Validar columna escalada

EmpleadosAttrition['MonthlyIncome_scaled'].describe()
```

```python
#Eliminar columna original MonthlyIncome

EmpleadosAttrition = EmpleadosAttrition.drop(['MonthlyIncome'], axis=1)
```

```python
# Convertir variables categóricas a numéricas usando Label Encoding

from sklearn.preprocessing import LabelEncoder

# Crear un objeto LabelEncoder
label_encoder = preprocessing.LabelEncoder()

# Crear bucle for para cada columna que se desea transformar con LabelEncoder

for column in ['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Attrition','OverTime']:
    EmpleadosAttrition[column] = label_encoder.fit_transform(EmpleadosAttrition[column])

# Verificar los cambios
EmpleadosAttrition.head()
```

```python
# Validación final de estructura del Frame

EmpleadosAttrition.info()
EmpleadosAttrition.describe()
```

```python

# Calcular la correlación de todas las variables con Attrition
correlation_with_attrition = EmpleadosAttrition.corr()['Attrition'].sort_values(ascending=False)

# Filtrar solo variables donde el valor absoluto del coeficiente de correlacion sea mayor o igual a 0.1
correlation_with_attrition_best = correlation_with_attrition[abs(correlation_with_attrition) >= 0.1]

# Mostrar las correlaciones significativas (mayor a 0.1)
correlation_with_attrition_best
```

```python
# Extrar del dataframe original un nuevo dataframe con solo las columnas que tiene una correlacion significativa con Attrition

# Obtener nombres de columnas
significant_columns = correlation_with_attrition_best.index.tolist()
significant_columns

# Usar los nombres de las columnas para crear el nuevo dataframe

EmpleadosAttritionFinal = EmpleadosAttrition[significant_columns]
EmpleadosAttritionFinal.info()

```

```python
from sklearn.decomposition import PCA

# Crear un objeto PCA.
pca = PCA()

# Ajustar el modelo PCA a los datos en EmpleadosAttritionFinal (excluyendo 'Attrition')
EmpleadosAttritionPCA = pca.fit_transform(EmpleadosAttritionFinal.drop('Attrition', axis=1))
```

```python
# Calcular y mostrar la varianza explicada por cada componente principal
variance_explained_percentage = pca.explained_variance_ratio_ * 100

# Formatear la salida con 3 decimales
formatted_variance = [f'{var:.3f}%' for var in variance_explained_percentage]

# Imprimir los resultados
print("Varianza explicada por cada componente principal:")
for i, var in enumerate(formatted_variance):
    print(f"Componente {i}: {var}")
```

```python
# Calcular la varianza explicada acumulada
Porcentaje_Varianza_Explicada_Acum = np.cumsum(pca.explained_variance_ratio_)

# Identificar componentes que explican al menos el 80% de la varianza
componentes_80 = np.argmax(Porcentaje_Varianza_Explicada_Acum >= 0.80) + 1

print(f"Número de componentes que explican el 80% de la varianza: {componentes_80}")
```

```python
# Crear un DataFrame con las componentes principales seleccionadas
pca_df = pd.DataFrame(EmpleadosAttritionPCA[:, :componentes_80],
                      columns=[f'C{i}' for i in range(componentes_80)])

# Agregar las componentes principales a EmpleadosAttritionFinal usando assign
EmpleadosAttritionFinal = EmpleadosAttritionFinal.assign(**pca_df)

EmpleadosAttritionFinal.head()
```

```python
# Guardar resultados

EmpleadosAttritionFinal.to_csv('EmpleadosAttritionFinal.csv', index=False)
```
